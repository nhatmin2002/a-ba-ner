```python
import pandas as pd
import numpy as np
with open(r'documents\covid\dev_syllable.text','r',encoding='utf8') as f:
    lines = f.readlines()
    lines = list(map(lambda x:x.strip(),lines))
index = np.arange(len(lines))

item = {"index":index,"text":lines}
df1 = pd.DataFrame(item)

df2 = pd.read_csv("data_withtext/covid/dev_syllable.csv")
df2.columns =['index', 'text', 'image', 'link']
main = pd.merge(df1, df2, left_on='index', right_on='index', how='left')

def filter_fn(row):
    if pd.isnull(row['image']):
        return False
    else:
        return True

m = main.apply(filter_fn, axis=1)
a = list(m)
data = []
import json
with open(r'raw_doc\Covid19\dev_word.json',encoding='utf8') as f:
    for line in f:
        words = json.loads(line)['words']
        tags = json.loads(line)['tags']
        data.append(list(zip(words,tags)))

filtered_list = [i for (i, v) in zip(data, a) if v]
image = list(df2['image'])
with open("data_ner_covid/dev",'w',encoding='utf8') as f:
    for item in list(zip(filtered_list,image)):
        ner_tag,img = item[0],item[1]
        f.write("IMGID:"+img.split('.')[0])
        f.write('\n')
        for word in ner_tag:
            f.write(word[0]+"\t"+word[1])
            f.write('\n')
        f.write("\n")
```
